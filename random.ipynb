```python
# Numerical arrays and random number.
import numpy as np

# Plot.
import matplotlib.pyplot as plt
```

## Generate Random Intergers

***

```python
rng = np.random.default_rng()
```

```python
# Generate one random float uniformly distributed over the range [0, 1)
rng.random()  
```

```python
# Generate an array of 10 numbers according to a unit Gaussian distribution.
rng.standard_normal(10)  
```

```python
# Generate an array of 5 integers uniformly over the range [0, 10).
rng.integers(low=0, high=10, size=5)  
```

```python
# Generate 100 intergers over the range [0, 10).
x = rng.integers(low=0, high=10, size=100, endpoint=True)

# Show.
x
```

```python
# Get the counts.
values, counts = np.unique (x, return_counts=True)

# Show.
values, counts
```

```python
# Plot the bar chart.
plt.bar(values, counts, edgecolor='black')
```

## Uniform Distribution

***

```python
# Generate 100 floats uniformly distributed over the range [0, 10).
x = rng.uniform(low=0, high=10, size=100)

# Show.
x
```

```python
# Get the counts.
values, counts = np.unique (x, return_counts=True)

# Show.
values, counts, values.size, counts.size
```

```python
# Plot the bar chart.
plt.bar(values, counts, edgecolor='black')
```

```python
# Round values in x.
x_rounded = np.round(x, decimals=0)

# Get the counts.
values, counts = np.unique (x_rounded, return_counts=True)

# Show.
values, counts, values.size, counts.size
```

```python
# Plot the bar chart.
plt.bar(values, counts, edgecolor='black')
```

```python
# Generate 100 floats uniformly distributed over the range [0, 10).
x = rng.uniform(low=0, high=10, size=10000)

# Show.
x
```

```python
# Get the counts.
values, counts = np.unique (x, return_counts=True)

# Show.
values, counts, values.size, counts.size
```

```python
# Plot histogram.
plt.hist(x, edgecolor='black')
```
